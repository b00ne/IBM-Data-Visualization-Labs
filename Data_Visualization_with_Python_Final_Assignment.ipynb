```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://cocl.us/datascience_survey_data', index_col=0)
df.head()
```

```python
df_sorted = df.sort_values(by='Very interested', ascending=False)

total_respondents = 2233

df_percent = df_sorted/total_respondents

df_percent.head()
```

```python
import matplotlib.patches as patches

bar_width = 0.8
color_list = ('#5cb85c', '#5bc0de', '#d9534f')

ax = df_percent.plot(kind='bar', figsize=(20, 8), width=bar_width, color=color_list)

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.2%}'.format(height), (x, y + height + 0.01), size=14)
    
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.tick_params(axis='x', which='major', labelsize=14)
ax.set_yticks([])
ax.legend(fontsize=14)
ax.set_title("Percentage of Respondents' Interest in Data Science Areas", size=16)
```

```python
df = pd.read_csv('https://cocl.us/sanfran_crime_dataset', index_col=0)
df.head(10)
```

```python
df.info()
```

```python
df.rename(columns={'PdDistrict':'Neighborhood'}, inplace = True)

df_n = df.groupby('Neighborhood').count()
df_n.reset_index(inplace=True)
df_n = df_n[['Neighborhood','X']]
df_n.rename(columns={'X':'Count'}, inplace=True)
df_n
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium

print('Folium installed and imported!')
```

```python
!wget --quiet https://cocl.us/sanfran_geojson -O sanfran.geojson
print('GeoJSON file downloaded!')
```

```python
sf_geo = r'sanfran.geojson'

sf_map = folium.Map(location=[37.77, -122.42], zoom_start=12, tiles='OpenStreetMap', width=800, height=600)
```

```python
sf_map.choropleth(
    geo_data=sf_geo,
    data=df_n,
    columns=['Neighborhood','Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Fransico'
)

sf_map
```
